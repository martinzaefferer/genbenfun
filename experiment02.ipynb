## Analysis of results / generating heatmaps

```python
import os, sys, importlib.util
import matplotlib.pyplot as plt

MODULE_PATH = "bbob_disturbance_analysis.py"

spec = importlib.util.spec_from_file_location("bbob_disturbance_analysis", MODULE_PATH)
bbob = importlib.util.module_from_spec(spec)
sys.modules["bbob_disturbance_analysis"] = bbob
spec.loader.exec_module(bbob)

results_path_csv = runfile
df_prepped = bbob.prepare_results(runfile)


print(df_prepped)
```

```python
# Directory to store all plots
out_dir = "heatmaps"
import numpy as np

# Choose metrics to produce combined panels for (one figure per metric + problem)
metrics = ["true", "delta"]

# Preferred algorithm ordering for side-by-side comparison
preferred_order = ["DE", "CMA-ES", "Random"]

for metric in metrics:
    df_hm = bbob.select_xgap_metric(df_prepped, metric=metric)
    out_metric_dir = os.path.join(out_dir, metric)
    os.makedirs(out_metric_dir, exist_ok=True)

    for problem in df_prepped["problem"].unique():
        dim = int(df_prepped.loc[df_prepped["problem"] == problem, "dim"].iloc[0])
        # determine algorithms present and order them
        present = list(df_hm["algo"].unique())
        algos = [a for a in preferred_order if a in present] + [a for a in present if a not in preferred_order]
        if len(algos) == 0:
            continue

        # collect matrices for each algo; ensure they share the same grid
        mats = []
        skip = False
        for algo in algos:
            try:
                A, F, Z = bbob.make_matrix_xgap(df_hm, problem, dim, algo, value_col="xgap_metric", stat="median")
            except Exception as e:
                print(f"Skipping {problem} | {algo}: {e}")
                skip = True
                break
            mats.append((algo, A, F, Z))

        if skip or len(mats) == 0:
            continue

        baseA, baseF = mats[0][1], mats[0][2]
        for _, A, F, Z in mats:
            if not (np.array_equal(A, baseA) and np.array_equal(F, baseF)):
                print(f"Skipping {problem}: amplitude/frequency grids differ between algorithms")
                skip = True
                break
        if skip:
            continue

        # derive plotting centers/edges (match plot_heatmap behavior)
        amp_centers = baseA
        freq_centers = baseF
        x_centers = np.asarray(freq_centers)
        y_centers = np.asarray(amp_centers)
        x_edges = bbob._edges_from_centers(x_centers)
        y_edges = bbob._edges_from_centers(y_centers)

        # prepare data arrays and compute global vmin/vmax for consistent color scale
        cs = []
        for _, A, F, Z in mats:
            C = np.asarray(Z).T  # swap_axes=True in plot_heatmap
            cs.append(C)
        all_vals = np.concatenate([c.flatten() for c in cs if c.size > 0])
        if all_vals.size == 0:
            print(f"No data to plot for {problem} ({metric})")
            continue
        vmin = float(np.nanmin(all_vals))
        vmax = float(np.nanmax(all_vals))

        # create side-by-side subplots with shared color scale and single colorbar
        n = len(mats)
        fig, axes = plt.subplots(1, n, figsize=(5 * max(1, n), 5), constrained_layout=True)
        if n == 1:
            axes = [axes]

        mesh = None
        for i, (algo, A, F, Z) in enumerate(mats):
            ax = axes[i]
            C = np.asarray(Z).T
            mesh = ax.pcolormesh(x_edges, y_edges, C, shading="auto", vmin=vmin, vmax=vmax)
            ax.set_xticks(x_centers)
            ax.set_yticks(y_centers)
            ax.set_xlabel("frequency")
            if i == 0:
                ax.set_ylabel("amplitude")
            ax.set_title(f"{problem} | {algo}")
            ax.set_xticks(x_edges, minor=True)
            ax.set_yticks(y_edges, minor=True)
            ax.grid(which="minor", linewidth=0.5, alpha=0.3)

        # single colorbar for the figure
        if mesh is not None:
            cbar = fig.colorbar(mesh, ax=axes, orientation="vertical")
            cbar.set_label(f"best_xgap (median) [{metric}]")

        safe_problem = problem.replace("/", "_")
        fname = f"{safe_problem}_metric-{metric}_algorithms.png"
        fpath = os.path.join(out_metric_dir, fname)
        fig.savefig(fpath, dpi=160, bbox_inches="tight")
        plt.close(fig)
        print(f"Saved: {fpath}")

print(f"All heatmaps saved to {out_dir}")
```
